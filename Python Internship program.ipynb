## Create a basic calculator that can perform basic arithmetic operations such as addition,subtraction, multiplication, and division.using functions

```python


# Calling functions for addition, substraction, miltiplication, devision
def add(x, y):
    return x+y #returning the value of addition of x and y
def substract(x, y):
    return x-y #returning the value of substraction of x and y
def mul(x, y):
    return x*y #returning the value of multiplication of x and y
def div(x, y):
    if y!=0:
        return x/y  # checking id denominator value is zero, because we can not devide any number by zero
    else:
        return "Error: You can not devide any number by zero"
        exit()

# function to take user input and perform calculations
def calculator():
    while True:
        time.sleep(1)
        print("1.Addition")
        print("2.Substraction")
        print("3.Multiplication")
        print("4.Division")
        print("5.Exit")

        choice=input("Enter a choice from(1/2/3/4/5):")
        
        if choice=='5':
                print("Thank you for using calculator function...!, Visit again!")
                break
                
        if choice in ['1', '2', '3', '4']:
            if choice=='1':
                num1= eval(input("Enter the first number: "))
                num2= eval(input("Enter the second number: "))
                addition=add(num1, num2)
                time.sleep(1)
                print("The addition of number {} and {} is: {}" .format(num1, num2, round(addition, 2)))
            elif choice=='2':
                num1= eval(input("Enter the first number: "))
                num2= eval(input("Enter the second number: "))
                substraction=substract(num1, num2)
                time.sleep(1)
                print("The substraction of number {} and {} is: {}" .format(num1, num2, round(substraction, 2)))
            elif choice=='3':
                num1= eval(input("Enter the first number: "))
                num2= eval(input("Enter the second number: "))
                multiplication=mul(num1, num2)
                time.sleep(1)
                print("The multiplication of number {} and {} is: {}" .format(num1, num2, round(multiplication, 2)))
            elif choice=='4':
                num1= eval(input("Enter the first number: "))
                num2= eval(input("Enter the second number: "))
                division=div(num1, num2)
                time.sleep(1)
                print("The division of number {} and {} is: {}" .format(num1, num2, division))
            else:
                print("Invalid input option, Please enter a valid choice: (1/2/3/4/5).")

calculator()
```

# Task 4 : Tic Tac Toe Game
## Create a simple Tic Tac Toe game that can be played between two players or against the computer.


```python
i=0
j=0

def design_board(board):
    for row in board:
        print(" | ".join(row))
        print("-"*9)
        
def choose_winner(board, player):
    #check row and column
    for i in range(3):
        if all(board[i][j]== player for j in range(3)) or all(board[j][i] ==player for j in range(3)):
            return True
    # Check diagonals
    if all(board[i][j]== player for i in range(3)) or all(board[i][2-i]==player for i in range(3)):
        return True
    return False

def check_bord(board):
    return all(board[i][j] !=' ' for i in range(3) for j in range(3))

def valid_move(board, row, col):
    return 0<=row <3 and 0<=col<3 and board[row][col] ==' '

def TTT_game():
    board=[[' ' for _ in range(3)] for _ in range(3)]
    current_player='X'
    while True:
        design_board(board)
        row= int(input("Player {}, Enter the row (0, 1, 2): " .format(current_player)))
        col= int(input("Player {}, Enter the row (0, 1, 2): " .format(current_player)))
        if not valid_move(board, row, col):
            print("Invalide move, please try again...!")
            continue
        board[row][col]=current_player
        if choose_winner(board, current_player):
            design_board(board)
            print("Player {} wins!!" .format(current_player))
            break
        if check_bord(board):
            design_board(board)
            print("Its a tie!")
            break
        current_player ='O' if current_player=='X' else 'X'


TTT_game()
```
